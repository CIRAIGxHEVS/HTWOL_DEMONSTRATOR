```python
import streamlit as st
import numpy as np
import matplotlib.pyplot as plt
import model_H2 as mh
```

```python
# Initialisation des paramètres
param_cell = [0.02]
param_stack = [20000, 10000, 1.3, 1]
param_cons = [0.85, 9, 14]
param_bop = [80000, ['BoP', 'Compressor'], [20, 12]]

kg_per_FU = 1

param_he = [397 * 1e-6, 'NiO-YSZ']
param_oe = [12 * 1e-6, 'LSM']
param_bl = [5 * 1e-6, 'LSMC']
param_cl = [100 * 1e-6, 'CGO']
param_el = [5 * 1e-6, 'YSZ8']
param_bp = [1000 * 1e-6, 'Steel']
param_seal = [200000 * 1e-6, 'glass mat']
param_ep = [3000 * 1e-6, 'Steel']

component = [param_he, param_oe, param_bl, param_cl, param_el, param_bp, param_seal, param_ep]

e_cell = component[:6]
e_stack = component[6:]

elec_mix = [0, 1, 0, 0]
heat_mix = [0.5, 0, 0.5]

tier = 2

# Interface utilisateur Streamlit
st.title("Simulation de l'Impact Environnemental")

# Option de modification des paramètres dans l'interface
kg_per_FU = st.slider("Quantité de kg par FU", min_value=1, max_value=10, value=kg_per_FU)

param_cell[0] = st.slider("Paramètre Cell", min_value=0.01, max_value=0.05, value=param_cell[0])

# Calcul de l'impact environnemental
result = mh.total_impact(kg_per_FU, param_cell, e_cell, param_stack, e_stack, param_bop, param_cons, elec_mix, heat_mix)[tier]

# Afficher les résultats
st.write("Résultat du calcul d'impact environnemental :", result)

# Affichage du graphique
fig = plt.figure()
mh.plotting(result, tier)  # Cette fonction crée un graphique avec matplotlib
st.pyplot(fig)
```
